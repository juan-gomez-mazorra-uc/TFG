```python
import matplotlib.pyplot as plt
import numpy as np

# Definir la función de potencial
def potential(x):
    if -2 <= x <= 2:
        return 0  # Dentro del pozo de potencial
    else:
        return 100  # Fuera del pozo de potencial

# Crear un rango de posiciones
x_values = np.linspace(-5, 5, 400)

# Calcular la energía potencial en cada posición
potential_values = [potential(x) for x in x_values]

# Crear el gráfico
plt.plot(x_values, potential_values, label='Energía Potencial', color='blue')

# Añadir etiquetas y leyenda
plt.xlabel('Posición')
plt.ylabel('Energía')
plt.ylim(0, 10)
plt.title('Energía Potencial en un Pozo Cuadrado')
# plt.legend()

# Mostrar el gráfico
# plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Definir los parámetros del pozo de potencial
L = 4  # Anchura del pozo
a = -L / 2
b = L / 2

# Definir la función de onda
def wave_function(x, n, L):
    return np.sqrt(2 / L) * np.sin(n * np.pi * (x - a) / L)

# Crear un rango de posiciones
x_values = np.linspace(a, b, 400)

# Calcular las funciones de onda para n=1 y n=2
psi_1 = wave_function(x_values, 1, L)
psi_2 = wave_function(x_values, 2, L)

# Crear el gráfico del pozo de potencial
plt.plot(x_values, np.zeros_like(x_values), 'blue', label='Energía Potencial')

# Añadir las funciones de onda
plt.plot(x_values, psi_1, 'r-', label='$\psi_1(x)$')
plt.plot(x_values, psi_2, 'g-', label='$\psi_2(x)$')

# Añadir líneas verticales para el infinito
plt.axvline(x=a, color='blue', linestyle='--')
plt.axvline(x=b, color='blue', linestyle='--')

# Añadir etiquetas y leyenda
plt.xlabel('Posición')
plt.ylabel('Amplitud de la Función de Onda')
plt.title('Funciones de Onda en un Pozo de Potencial Cuadrado Infinito')
plt.legend()

# Mostrar el gráfico
# plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Definir los parámetros del pozo de potencial
L = 2  # Anchura del pozo
a = 0
b = L

# Definir la función de onda
def wave_function(x, n, L):
    return np.sqrt(2 / L) * np.sin(n * np.pi * (x - a) / L)

# Definir las energías de los niveles
def energy_level(n, L):
    return (n**2 * np.pi**2) / (2 * L**2)

# Crear un rango de posiciones
x_values = np.linspace(a, b, 400)

# Calcular las funciones de onda y las energías para n=1, n=2, y n=3
psi_1 = wave_function(x_values, 1, L)
psi_2 = wave_function(x_values, 2, L)
psi_3 = wave_function(x_values, 3, L)

energy_1 = energy_level(1, L)
energy_2 = energy_level(2, L)
energy_3 = energy_level(3, L)

# Crear el gráfico del pozo de potencial
plt.plot(x_values, np.zeros_like(x_values), 'blue', label='Energía Potencial')

# Añadir las funciones de onda desplazadas por sus energías
plt.plot(x_values, psi_1 + energy_1, 'r-', label='$\psi_1(x)$ (n=1)')
plt.plot(x_values, psi_2 + energy_2, 'g-', label='$\psi_2(x)$ (n=2)')
plt.plot(x_values, psi_3 + energy_3, 'b-', label='$\psi_3(x)$ (n=3)')

# Añadir líneas verticales para el infinito
plt.axvline(x=a, color='black', linestyle='-')
plt.axvline(x=b, color='black', linestyle='-')

# Añadir etiquetas y leyenda
plt.xlabel('$x$')
plt.ylabel('$E$')
plt.ylim(0, 20)
plt.title('Funciones de Onda y Energías en un Pozo de Potencial Cuadrado Infinito')
plt.legend()

# Mostrar el gráfico
# plt.grid(True)
plt.show()
```
