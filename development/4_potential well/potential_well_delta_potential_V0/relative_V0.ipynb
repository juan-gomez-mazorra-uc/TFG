```python
import os
import csv
import numpy as np
import matplotlib.pyplot as plt

v_0 = []
energy = []

folder = 'results_COBYLA_N=2'
files_csv = ['shots=16384.csv']


for index, archivo_csv in enumerate(files_csv):

    v_0.append([])
    energy.append([])

    with open(os.path.join(folder, archivo_csv), 'r') as file:
        reader = csv.reader(file)
        next(reader)
        for row in reader:
            v_0[index].append(float(row[0]))
            energy[index].append(float(row[1]))

files_names = [file.split('.')[0] for file in files_csv]
```

```python
files_names
```

```python
# Quitamos v_0 para que al dividir no sea infinito
v_0 = v_0[0][1:]
energy = energy[0][1:]
```

```python
for i in range(len(v_0)):
    energy[i] = (energy[i] - 9.869604401089362) / v_0[i]
```

```python
plt.plot(v_0, energy, marker='', linestyle='-', label=files_names[0])
plt.xlabel('v_0')
plt.ylabel('(Energy -9.869604401089362) / v_0')
plt.title('Delta Potential')
plt.legend()
plt.savefig(f'{folder}/delta_potential_relative_v0')
plt.show()
```
